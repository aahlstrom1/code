A demonstration of using DESI mass spectrometry data, with some analysis and visualizations

```python
import h5py
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import random
import scipy.io as sio
from scipy.spatial import distance_matrix
import warnings; warnings.simplefilter('ignore')
```

```python
# Read in the data from the labeled spectrum file
file_name = '../../Labeled DESI Data/8-27-18 A3D16F3 slide 2 brain 1.mat'
f = h5py.File(file_name, 'r')
data = {}
for k, v in f.items():
    data[k] = np.array(v)
print (f.items())
```

```python
# View the titles of the data sets included in the file
[(k,v) for k,v in f.items()]
```

```python
# Get the regions from the labeled data file
region_count = int(data["ROIcount"])
regions = [np.array([x for x in f[data["ROI"][1,i]]]) for i in range(region_count)]

# Display the brain by displaying intensities at the m/z value that maximizes
# the intensity at a pixel within the body of the brain
pixel = (20,20)
brain_image = data['img'][np.argmax(data['img'][:,pixel[0],pixel[1]]),:,:]
print("Brain images at m/z value of",
        data["imgZ"][0][np.argmax(data['img'][:,pixel[0],pixel[1]])])
plt.subplot(121).imshow(brain_image)
plt.subplot(121).set_title("Brain without\nregions highlighted")
plt.subplot(122).imshow(brain_image + np.multiply(brain_image, sum(regions)))
plt.subplot(122).set_title("Brain with\nregions highlighted")
plt.show()

# Initialize containers for the region names and assigned colors
region_names = []
colors = ['k', 'r', 'y', 'g', 'b']

# Determine region names and amount of pixels in each region
print("Pixel counts in regions:")
for i in range(region_count):
    region_names.append(''.join([chr(x[0]) for x in f[data["ROI"][0,i]]]))
    print(region_names[-1], end=": ")
    print(sum(sum(regions[i])), "pixels in region")

# Output the total amount of pixels in the image
shape = data["img"].shape
print("\nTotal pixels in image:", shape[1]*shape[2])
```

```python
# Define the amount of highest peaks to identify
selection_size = 100

# Initialize containers for the relevant output
xSize, ySize, zSize= data['img'].shape
spectra, current = np.zeros((ySize*zSize, xSize)), 0
maxes = []
aggregated_maxes = [{} for r in range(region_count)]
region_point_counts = [0]*region_count

# Iterate through all pixels in the given image
for i in range(ySize):
    for j in range(zSize):
        # Sort the intensity values to find the 100 highest
        spectra[current], current = data['img'][:,i,j], current + 1
        sort = np.argsort(spectra[current-1])[-selection_size:]
        maxes.append({data["imgZ"][0][k]:spectra[current-1][k] for k in sort})
        
        # If this pixel is in a region of interest, plot and save the data
        for r in range(region_count):
            if regions[r][i,j] == 1:
                keys, values = ([k for k in maxes[-1].keys()],
                        [v for v in maxes[-1].values()])
                for k in keys:
                    aggregated_maxes[r][k] = ([] if k not in aggregated_maxes[r]
                            else aggregated_maxes[r][k]) + [maxes[-1][k]]
                plt.subplot(711+r).plot(keys, values, colors[r]+'.')
                region_point_counts[r] += 1
        
        # An arbitrary point outside the brain
        if i == 40 and j == 2:
            keys, values = ([k for k in maxes[-1].keys()],
                    [v for v in maxes[-1].values()])
            point_maxes = {}
            plt.subplot(716).plot(keys, values, 'c.')
            
        # An arbitrary point inside the brain, but outside regions of interest
        if i == 75 and j == 30:
            keys, values = ([k for k in maxes[-1].keys()],
                    [v for v in maxes[-1].values()])
            point_maxes = {}
            plt.subplot(717).plot(keys, values, 'm.')
            
        # Progress output
        if current % 100 == 0:
            print('\r' + str(current), 'out of', ySize * zSize, end='')

# Plot the resulting data
for r in range(7):
    plt.subplot(711+r).set_xlim(250, 900)
    plt.subplot(711+r).set_ylim(0, 125000)
plt.subplot(711).set_title("Top 100 Points for Every Spectrum in Each Region")
plt.subplot(717).set_xlabel("Molecular Size (m/z)")
plt.subplot(714).set_ylabel("Intensity (ion count)")
plt.show()
```

```python
# Define the amount of highest peaks to identify
selection_size = 500

# Initialize containers for the relevant output
xSize, ySize, zSize= data['img'].shape
spectra, current = np.zeros((ySize*zSize, xSize)), 0
maxes = []
aggregated_maxes = [{} for r in range(region_count)]
region_point_counts = [0]*region_count

# Iterate through all pixels in the given image
for i in range(ySize):
    for j in range(zSize):
        # Sort the intensity values to find the 100 highest
        spectra[current], current = data['img'][:,i,j], current + 1
        sort = np.argsort(spectra[current-1])[-selection_size:]
        maxes.append({data["imgZ"][0][k]:spectra[current-1][k] for k in sort})
        
        # If this pixel is in a region of interest, plot and save the data
        for r in range(region_count):
            if regions[r][i,j] == 1:
                keys, values = ([k for k in maxes[-1].keys()],
                        [v for v in maxes[-1].values()])
                for k in keys:
                    aggregated_maxes[r][k] = ([] if k not in aggregated_maxes[r]
                            else aggregated_maxes[r][k]) + [maxes[-1][k]]
                plt.subplot(511+r).plot(keys, values, colors[r]+'.')
                region_point_counts[r] += 1
                
        # Progress output
        if current % 100 == 0:
            print('\r' + str(current), 'out of', ySize * zSize, end='')

# Plot the resulting data
for r in range(region_count):
    plt.subplot(511+r).set_xlim(250, 900)
    plt.subplot(511+r).set_ylim(0, 125000)
plt.subplot(511).set_title("Top 500 Points for Every Spectrum in Each Region")
plt.subplot(515).set_xlabel("Molecular Size (m/z)")
plt.subplot(513).set_ylabel("Intensity (ion count)")
plt.show()
```

```python
# Calculate mean values
means = [{k:sum(aggregated_maxes[r][k])/region_point_counts[r]
        for k in aggregated_maxes[r]} for r in range(region_count)]

# Graph the mean values across the entire m/z axis
for r in range(region_count):
    plt.plot(means[r].keys(), means[r].values(), colors[r]+'.')
plt.title("Mean Intensity Values for Each of the Five Regions at Each\n" +
        "m/z Value at which a Top-500 Intensity was Achieved")
plt.xlabel("Molecular Size (m/z)")
plt.ylabel("Intensity (ion count)")
plt.show()

# Graph the mean values across the 833 to 840 m/z range
for r in range(region_count):
    plt.plot(means[r].keys(), means[r].values(), colors[r]+'.')
plt.xlim(833, 840)
plt.title("Mean Intensities Restricted to the 833-840 m/z Range")
plt.xlabel("Molecular Size (m/z)")
plt.ylabel("Intensity (ion count)")
plt.show()

# Graph the mean values across the 834.3 to 834.8 m/z range
for r in range(region_count):
    sort = np.argsort(list(means[r].keys()))
    plt.plot(np.array(list(means[r].keys()))[sort], np.array(list(means[r].values()))[sort],
            colors[r]+'.-', label=region_names[r])
plt.xlim(834.3, 834.8)
plt.legend()
plt.title("Mean Intensities Restricted to the 834.3-834.8 m/z Range")
plt.xlabel("Molecular mass-to-charge (m/z)")
plt.ylabel("Intensity (ion count)")
plt.show()
```

```python
# Fit a normal curve to the data
highest_peak = max(means[r], key=means[r].get)
sigmas = [np.sqrt(np.var(aggregated_maxes[r][highest_peak]))
        for r in range(region_count)]
mus = [means[r][highest_peak] for r in range(region_count)]
subplots = [322+i for i in range(region_count)]

# Plot the normal curves together
domain = np.linspace(0, 130000, 10000)
for r in range(region_count):
    plt.subplot(321).plot(domain,mlab.normpdf(domain,
            mus[r], sigmas[r]), colors[r]+'-')
plt.subplot(321).set_title((" "*46) + "Normalized Region " +
        "Histograms with Fitted Normal Curves")

# Plot each of the regions' normal curves with intensity histograms
for r in range(region_count):
    maxes = aggregated_maxes[r][highest_peak]
    domain = np.linspace(min(maxes), max(maxes), 10000)
    plt.subplot(subplots[r]).hist(maxes, normed=True)
    plt.subplot(subplots[r]).plot(domain,mlab.normpdf(domain,
            mus[r], sigmas[r]), colors[r]+'-')
    if subplots[r] >= 325:
        plt.subplot(subplots[r]).set_xlabel("Intensity")
    elif subplots[r] == 323:
        plt.subplot(subplots[r]).set_ylabel("Normalized Frequency")

plt.show()
```
