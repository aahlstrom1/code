The data yielded from Bruker and Agilent are different. Bruker gives intensities at about 60,000 different m/z values, Agilent at about 100,000. This is a preliminary test of standardizing such data to a single range of m/z values.

The end of this file uses output from the "SVD Projection" file.

Note that both this file and "SVD Projection" are somewhat memory intensive; I only get them to run when they're the only notebook running. I haven't even been using this stuff recently, anyways; it's just sort of documenting things that I experimented with.

```python
import gc
import h5py
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import random
import scipy.io as sio
import scipy.sparse as ssp
import warnings; warnings.simplefilter('ignore')
```

```python
# Define initial parameters, including Pandas DataFrame column names
directory = '../../DESI Data'
images = []
file_properties = {
    'File':[],
    'Z-axis':[],
    'Standardized Z-axis':[],
    'X_size': [],
    'Y_size': [],
    'Z_size': []
}
spectra = {
    'X_coordinate': [],
    'Y_coordinate': [],
    'Original Spectrum': [],
    'Mean-adjusted Spectrum': [],
    'Standardized Spectrum': [],
    'Source File': []
}

# Define the parameters for resampling
new_interval_count = int(1e5)
new_min, new_max = 100, 1000
new_z_axis = np.linspace(new_min, new_max, new_interval_count+1)

# Iterate through all files in the given directory
for file in os.listdir(directory):
    # Initialize variables, including container for the 3-d tensor (image)
    image, z_axis = None, None
    x, y, z = None, None, None
    
    try:
        # The file.bin file has been causing problems.
        # Maybe replace or get rid of it?
        if file == "file.bin":
            print("Skipping", file)
            continue
        
        # Handle standard binary files (Agilent)
        elif file.endswith('.bin'):
            f = open(directory + '/' + file, 'rb')
            data = np.fromfile(f, '<f4')
            x, y, z = [int(n) for n in data[:3]]
            image, z_axis = data[3:x*y*z + 3], data[-z:]
            image.shape = (y, x, z)
            image = image.transpose((1, 0, 2))

        # Handle MATLAB binary files (Bruker)
        elif file.endswith('.mat'):
            data = {}
            f = h5py.File(directory + '/' + file, 'r')
            for k, v in f.items():
                data[k] = np.array(v)
            image, z_axis = data['img'], data['imgZ'][0]
            x, y, z = [len((data[x] if x != 'imgZ' else data[x].T))
                    for x in ['imgX', 'imgY', 'imgZ']]
            image.shape = (z, x, y)
            image = image.transpose((1, 2, 0))

    except:
        print("Unable to open", file)
        continue
    
    # Add relevant parameters to dict, later to be converted to Pandas DataFrame
    file_properties['File'].append(file)
    file_properties['Z-axis'].append(z_axis)
    file_properties['Standardized Z-axis'].append(new_z_axis)
    file_properties['X_size'].append(x)
    file_properties['Y_size'].append(y)
    file_properties['Z_size'].append(z)
    
    # Initialize sparse matrix and calculate resampled m/z axis
    conversion_matrix = ssp.lil_matrix((new_interval_count+1, z))
    min_mask = np.append(z_axis[1:], np.inf) >= new_min
    max_mask = np.append(-np.inf, z_axis[min_mask][:-1]) <= new_max
    z_axis = z_axis[min_mask][max_mask]
    
    # Add the image to the data, and calculate the mean-adjusted image
    images.append(image)
    int_image = image.astype(int)
    mean_adjusted_image = image/np.mean(image)
    
    # Build the sparse matrix to perform resampling on the data array
    index = 0
    for z in new_z_axis:
        right = len(z_axis[z_axis <= z])
        left = right - 1
        left_dist, right_dist = z - z_axis[left], z_axis[right] - z
        conversion_matrix[index,left] = right_dist/(left_dist + right_dist)
        conversion_matrix[index,right] = left_dist/(left_dist + right_dist)
        index += 1
    conversion_matrix = ssp.csr_matrix(conversion_matrix)
    random_index = np.random.randint(x*y)

    # Iterate through each pixel in the image, by row and then column
    index = 0
    for i in range(x):
        for j in range(y):
            # Compute resampled spectrum, and add features to dict
            standardized_spectrum = conversion_matrix @ int_image[i][j]
            spectra['X_coordinate'].append(i+1)
            spectra['Y_coordinate'].append(j+1)
            spectra['Original Spectrum'].append(int_image[i][j])
            spectra['Mean-adjusted Spectrum'].append(mean_adjusted_image[i][j])
            spectra['Standardized Spectrum'].append(standardized_spectrum)
            spectra['Source File'].append(file)
            
            # Plot original and resampled version of a random spectrum
            if index == random_index:
                plt.subplot(131).plot((z_axis if file != 'brain_2.bin' else z_axis[:-18]),
                        (int_image[i][j][min_mask][max_mask] if (file != 'brain_2.bin') 
                        else int_image[i][j][min_mask][max_mask][:-18]))
                plt.subplot(131).set_xlabel("Molecular Size (m/z)")
                plt.subplot(131).set_ylabel("Abundance (ion count)")
                plt.subplot(131).set_title("Example Original Spectrum")
                plt.subplot(133).plot(new_z_axis, conversion_matrix @ int_image[i][j])
                plt.subplot(133).set_xlabel("Molecular Size (m/z)")
                plt.subplot(133).set_ylabel("Abundance (ion count)")
                plt.subplot(133).set_title("Example Standardized Spectrum")
                plt.show()
            
            index += 1
```

```python
# Save the data to Pandas DataFrames
spectra_data_frame = pd.DataFrame(data=spectra)
file_properties_data_frame = pd.DataFrame(data=file_properties)
```

```python
# Add the SVD projections calculated in separate file
projection = np.load('projection.npz')['arr_0'].T
spectra_data_frame['Projected Spectrum'] = pd.Series([p for p in projection],
        index=spectra_data_frame.index)
```
